```python
new_content = []
for i in data["content"]:
    a = re.sub("\n","",i)
    a = re.sub("\u3000","",a)
    a = re.sub(" ","",a)
    a = re.sub("\uf67e","",a)
    a = re.sub("\uf67d","",a)
    a = re.sub("\uf67c","",a)
    new_content.append(a)
data["content"] = new_content
data["content"][0]
```

```python
new_content = []
for i in range(0,100):
    times = len(re.findall("[一二三四五六七八九十]",data["content"][i]))
    for e in range(times):
        clean = data["content"][i]
        clean = re.sub("一","1",clean)
        clean = re.sub("二","2",clean)
        clean = re.sub("三","3",clean)
        clean = re.sub("四","4",clean)
        clean = re.sub("五","5",clean)
        clean = re.sub("六","6",clean)
        clean = re.sub("七","7",clean)
        clean = re.sub("八","8",clean)
        clean = re.sub("九","9",clean)
        clean = re.sub("百","",clean)
        clean = re.sub("十","",clean)
    new_content.append(clean)
data["content"] = new_content
```

```python
code = []
for i in range(0,100):
    code.append(re.findall(r"[刑民]法第\d+條",data["content"][i]))
data["code"] = code
```

```python
code_count = {}
for i in range(50):
    for e in range(len(data["code"][i])):
        try:
            code_count[data["code"][i][e]] = code_count[data["code"][i][e]] + 1
        except:
            code_count[data["code"][i][e]] = 1
```

```python
price = []
for i in range(0,100):
    money = re.findall("[(賠償)(給付)(共計)].{0,10}\d萬?\d*,?\d+元",data["content"][i])
    try:
        money = re.findall("\d*,?\d*,?\d萬?\d*,?\d+",money[0])
        money[0] = re.sub("萬","",money[0])
        price.append(money[0])
    except:
        price.append(0)
data["price"] = price
```

```python
new_content = []
for i in range(0,100):
    times = re.findall("\d+成",data["content"][i])
    compile_re = re.compile("(\d+)(成)")
    for e in range(len(times)):
        times_ = compile_re.sub(r"\1 0%",data["content"][i])
    try:
        times_ = re.split(" ",times_)
        clean_data = ""
        for i in range (len(times_)):
            if times_[i] == " ":
                times_[i] = ""
            clean_data = clean_data + times_[i]
        new_content.append(clean_data)
    except:
        new_content.append(data["content"][i])
data["content"] = new_content
```

```python
data
```

```python
rate = []
for i in range(0,100):
    percent = re.findall("[故判本院庭決議依法得可見查心證].{0,100}被告.{0,8}[負擔].{0,9}\d+[%％]",data["content"][i])
    other = re.findall("[故判本院庭決議依法得可見查心證].{0,100}\d+[%％].{0,10}被告負擔",data["content"][i])
    for f in range(len(other)):
        percent.append(other[f])
            
    try:
        percent_ = re.findall("被告.{0,9}[負擔].{0,9}\d+[%％]",percent[-1])
        other = re.findall("\d+[%％].{0,10}被告負擔",percent[-1])
        for f in range(len(other)):
            percent_.append(other[f])
        percent = re.findall("\d+[%％]",percent_[-1])
        rate.append(percent[-1])
    except:
        rate.append(0)
data["rate"] = rate
```
