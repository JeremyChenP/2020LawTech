```python
new_content = []
for i in data["content"]:
    a = re.sub("\n","",i)
    a = re.sub("\u3000","",a)
    a = re.sub(" ","",a)
    a = re.sub("\uf67e","",a)
    a = re.sub("\uf67d","",a)
    a = re.sub("\uf67c","",a)
    new_content.append(a)
data["content"] = new_content
data["content"][0]
```

```python
code = []
for i in range(0,50):
    code.append(re.findall(r"[刑民]法第[\d一二三四五六七八九百十]+條",data["content"][i]))
data["code"] = code
```

```python
clean_code = []
for i in range(0,50):
    each_clean = []
    for e in range(len(data["code"][i])):
        clean = re.sub("一","1",data["code"][i][e])
        clean = re.sub("二","2",clean)
        clean = re.sub("三","3",clean)
        clean = re.sub("四","4",clean)
        clean = re.sub("五","5",clean)
        clean = re.sub("六","6",clean)
        clean = re.sub("七","7",clean)
        clean = re.sub("八","8",clean)
        clean = re.sub("九","9",clean)
        clean = re.sub("[百十]","",clean)
        each_clean.append(clean)
    clean_code.append(each_clean)
data["code"] = clean_code
```

```python
code_count = {}
for i in range(50):
    for e in range(len(data["code"][i])):
        try:
            code_count[data["code"][i][e]] = code_count[data["code"][i][e]] + 1
        except:
            code_count[data["code"][i][e]] = 1
```

```python
chi_dict = {
    "一" : 1,
    "二" : 2,
    "三" : 3,
    "四" : 4,
    "五" : 5,
    "六" : 6,
    "七" : 7,
    "八" : 8,
    "九" : 9,
    "十" : 10,
    "壹" : 1,
    "貳" : 2,
    "參" : 3,
    "肆" : 4,
    "伍" : 5,
    "陸" : 6,
    "柒" : 7,
    "捌" : 8,
    "玖" : 9,
    "百" : 100
}

def get_ratio_def(content):
    text_ratio = re.findall("被告[應負擔][^抗辯辯稱].{,20}[0-9一二三四無六七八九壹貳參肆伍陸柒捌玖].{,5}[成比例過失責任][^條元項]",content)
    temp = []
    for sentence in text_ratio:
        if ("％" in sentence) or ("%" in sentence):
            num_ratio = re.findall("\d.{,1}",sentence)[-1]
            temp.append(eval(num_ratio))
        elif ("成" in sentence):
            num_ratio = re.findall("[1-9一二三四五六七八九]",sentence)
            if num_ratio[-1].isdigit():
                temp.append(eval(num_ratio[-1])*10)
            else:
                temp.append(chi_dict[num_ratio[-1]]*10)
        elif ("分之" in sentence):
            match = re.search("(?P.)分之(?P.{,2})",sentence)
            try:
                if match.group('deno').isdigit():
                    num_ratio = eval(match.group('nume')[0])*10
                else:
                    num_ratio = eval(match.group('nume'))
                temp.append(num_ratio)
            except Exception as err:
                temp.append(80)
        else:
            temp.append(0)
    if temp:
        return temp[0]
    else:
        return 0
    

data['ratio'] = data['content'].apply(get_ratio_def)
```
